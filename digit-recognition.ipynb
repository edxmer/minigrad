```python
from engine import Value
from neural_network import MLP
import random
```

```python
with open("optdigits.tra", "r", encoding="utf-8") as f:
    lines = f.readlines()
    values_raw = [[int(x) for x in l.strip().split(',')] for l in lines]

class xy:
    def __init__(self, x, y):
        self.x = x; self.y = y
    def __repr__(self):
        return f'({self.x}, {self.y})'

inp_out = []
for v in values_raw:
    x = v[0:64]
    y = [0] * 10
    y[v[64]] = 1
    
    inp_out.append(xy(x, y))
 
```

```python
network = MLP(64, [16, 16, 16, 10])
```

```python
batch_size = 200
rounds = 200


chosen = random.choices(inp_out, k=batch_size)

loss = Value(0);

for t in chosen:
    act = network(t.x)
    for i in range(10):
        loss += (act[i]-t.y[i])**2

loss.backward()

for arg in network.arguments():
    arg.data -= arg.grad * 0.15

loss
```
