```python
from engine import Value
from neural_network import MLP
```

```python
a = Value(4)
b = Value(6)

def calc(a, b):
    c = a * b
    c -= a + 2
    d = c/a
    d **= 2
    a += d * 0.2 / c.relu()
    l = (a*b-c/4-d).tanh()
    return l

l0 = calc(a, b)

print('l0:\t', l0)

l0.backward()
print('a.grad:\t', a.grad)
print('b.grad:\t', b.grad)

a.data -= a.grad*0.01
b.data -= b.grad*0.01

l1 = calc(a, b)

print('l1:\t', l1)

```

```python
network = MLP(4, [4, 4, 1])

xs = [
    [1, 0.2, 0.4, .5],
    [-1, .4, -.4, -.1],
    [-.4, .3, .2, -.8],
    [.7, .7, .7, -.7]
]
ys = [1, -1, 1, -1]

def loss():
    return sum([(network(x)[0] - y)**2 for (x,y) in zip(xs, ys)])
```

```python
[network(x)[0] for x in xs]
```

```python
for i in range(1, 400):
    L0:Value = loss()
    print(f'{i:3}. loss:', L0)
    L0.backward()
    args = network.arguments()
    for argument in args:
        argument.data -= argument.grad * 0.01
```
