```python
from engine import Value
from neural_network import MLP
import random
from dataclasses import dataclass
import math
```

```python
# Creating synthetic dataset
# input-expected_output pairs

@dataclass
class inp_out:
    inp: list[float]
    out: float

def generate_test() -> inp_out:
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    is_in_circle = 1 if math.sqrt(x**2 + y**2) < 0.75 else -1
    return inp_out([x, y], is_in_circle)

data_size:float = 100
data = [generate_test() for _ in range(data_size)]
```

```python
network = MLP(2, [16, 16, 1], normalise=Value.relu, output_normalise=Value.tanh)
```

```python
k = 1.0
```

```python
rounds = 100

kend = k+rounds

while k < kend:
    loss = Value(0)
    for test in data:
        out = network(test.inp)[0]
        #print(test.out, out)
        d = (out-test.out)**2 / data_size
        #print(d)
        loss += d
    loss.backward()

    lr = 0.05
    for arg in network.arguments():
        arg.data -= arg.grad * lr
    
    print(f"{int(k):3}. round:", loss)
    k+=1
```

```python
print(network([0, 0]))
print(network([1, 1]))
print(network([1, 0]))
print(network([0, 1]))
print(network([-1, -1]))
print(network([.5, .5]))
```
